# boolean 으로 특정 row 조회하기

```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('./train.csv')
```

```python
train_data.head()
```

```python
# series 객체편에서 살펴봤듯이 boolean 형태로 리턴한다. 

pclass = train_data['Pclass'] == 1 
age = (train_data['Age'] >= 30) & (train_data['Age'] < 40)

pclass & age     # 조건 중첩도 가능
```

```python
train_data[pclass & age].head()  # 중첩된 조건으로 데이터프레임에 적용하면 필터링된 데이터프레임이 반환된다.
```

# DataFrame에서 새 컬럼 추가 혹은 컬럼 삭제하기

```python
# row 개수에 맞게 0에서 1사이의 랜덤 숫자 생성하기

rows = train_data.shape[0]
random_nums = np.random.rand(rows)
```

```python
# 새 칼럼 생성하기
train_data['random_nums'] = random_nums # 없는 이름으로 조회하려 하면 알아서 칼럼 생성 후 알아서 제일 끝에 붙여준다
# 원하는 위치에 넣고 싶으면 train_data.insert(num, col_name, col_value) 로 적용해도 된다. 
train_data.head()
```

```python
train_data['Fare'] = round(train_data['Fare'],2) # 원래 있던 값을 일괄변경하려면 iterable 하게 변경도 가능하다. 
train_data.head()
```

```python
# 칼럼 삭제하기 .drop(axis=1), 0인 경우는 row에 적용된다. 

train_data.drop('random_nums', axis=1, inplace=True)
train_data.head(2)
```

```python
# 번외: 조건에 해당하는 row 삭제하기

t2 = train_data[age & pclass]
t2_idx = t2.index
len(list(t2_idx))
```

```python
train_data.drop(t2_idx)   # 50개 빠지고 841 rows 남음
```
