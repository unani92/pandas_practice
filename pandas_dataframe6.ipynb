```python
import pandas as pd
import numpy as np

train_data = pd.read_csv('./train.csv')
```

# 형변환하기

```python
train_data.head()     # pclass는 생긴건 int이지만 범주형 데이터이다.
```

```python
# int에서 str로 형변환하기

train_data['Pclass'] = train_data['Pclass'].astype(str)
```

```python
# Age 변수 범주형으로 변환하기
# 변환 로직(함수)를 정의한 뒤 apply 함수로 적용한다.

import math

mean = round(train_data['Age'].mean())
def age_category(age):
    if math.isnan(age):    # 결측치가 있으면 apply 적용에 에러가 발생한다.
        return math.floor(mean/10) * 10
    else :
        return math.floor(age/10) * 10
```

```python
train_data['Age'].apply(age_category)
```

## group_by

```python
df = pd.read_csv('./train.csv')
```

```python
pclass = df.groupby('Pclass')
gender = df.groupby('Sex')
```

```python
pclass.groups         # 그룹 별 인덱스를 출력
```

```python
gender.count()      # min max, mean, sum 등을 사용할 수 있다.
```

```python
pclass_sex = df.groupby(['Pclass', 'Sex']).mean()   # 다중으로 그룹바이도 가능하다. 
```

```python
pclass_sex[['Survived','Age']]  # 그룹바이의 위력: 한큐에 좌석등급별, 성별 생존률을 알 수 있다. 
```

### 나이대별로 생존률 구하기

```python
import math

def age_category(age): 
    if math.isnan(age):
        return -1
    else :
        return f'{math.floor(age/10) * 10} 대'
```

```python
df['Age'] = df['Age'].apply(age_category)
```

```python
df.head()
```

```python
age = df.groupby('Age')
```

```python
a = age.mean()
type(a)
```

```python
age.count()
```

## aggregate

> 원하는 기초통계치 다 넣고 다 뽑아주는 기능

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0,1]).aggregate([np.mean, np.sum, np.max])
```
